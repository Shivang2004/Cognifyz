## Cuisine Combination

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv("C:/Users/Om/Downloads/tasks.csv")       # Loading the dataset
```

```python
df.head()
```

```python
df = df[['Cuisines', 'Aggregate rating']].dropna()    # Drop rows with missing cuisines and ratings
```

```python
common = df['Cuisines'].value_counts().head(10)       # Top 10 most common cuisines or ratings
```

```python
avg_rating = df.groupby('Cuisines')['Aggregate rating'].mean()
avg_rating = avg_rating.loc[common.index]                   
```

```python
plt.figure(figsize=(10, 5))  
plt.barh(common.index[::-1], common.values[::-1], color='skyblue')
plt.title('Top 10 Most Common Cuisine Combinations')
plt.xlabel('Number of Restaurants')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))        
plt.barh(avg_rating.index[::-1], avg_rating.values[::-1], color='orange')
plt.title('Average Rating of Top 10 Cuisine Combinations')
plt.xlabel('Average Rating')
plt.tight_layout()
plt.show()
```
