## Price Range vs. Online Delivery and Table Booking

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Om/Downloads/tasks.csv")                    # Loading the dataset
```

```python
df.head()
```

```python
df = df[['Price range', 'Has Online delivery', 'Has Table booking']]
```

```python
df['Online Delivery'] = df['Has Online delivery'].str.strip().str.lower().map({'yes':1, 'no':0})
df['Table Booking'] = df['Has Table booking'].str.strip().str.lower().map({'yes':1, 'no':0})
```

```python
grouped = df.groupby('Price range')[['Online Delivery', 'Table Booking']].mean()*100
```

```python
grouped.plot(kind='bar', figsize=(8, 5), color=['orange', 'blue'])
plt.title('Online Delivery & Table Booking by Price Range')                                # Plot distribution
plt.xlabel('Price Range')
plt.ylabel('Percentage of Restaurants')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title='Service')
plt.tight_layout()
plt.show()
```
